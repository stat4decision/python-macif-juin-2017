Importez la base de données SQLite salaries et récupérez la table Salaries dans un DataFrame 


```python
from sqlalchemy import create_engine
```

```python
engine=create_engine('sqlite:///salaries.sqlite')
```

```python
engine.table_names()
```

```python
#création du frame à partir du tableau SQL
frame_salaries=pd.read_sql_query("SELECT * from Salaries",engine)
```

```python
frame_salaries.shape
```

```python
frame_salaries.info()
```

```python
frame_salaries["TotalPay"].mean()
```

```python
frame_salaries.sort_values("TotalPay",ascending=False)
```

Joindre les deux dataframes de ventes en utilisant l’id.


```python
frame_merge=pd.merge(frame_ventes,frame_excel,how="inner",left_index=True,
                     right_on="account")
```

```python
frame_merge
```

### Discrétisation

```python
#nouvelle variable discrétisée
frame_merge["Jan_disc"]=pd.cut(frame_merge["Jan"],4)
```

```python
frame_merge["Jan_disc"]
#afficher toutes les modalités
frame_merge["Jan_disc"].unique()
```

```python
#fréquences par modalité
frame_merge["Jan_disc"].value_counts()
```

# Dataviz

```python
import matplotlib.pyplot as plt
```

```python
#optimisation des affichages sur notebook
%matplotlib inline
```

```python
# c= choix de la couleur des points
# s= taille des points
plt.scatter(frame_merge["Jan"],frame_merge["Feb"],c="red",
            s=frame_merge["Mar"]/1000)
plt.title("Janv vs Fev")
plt.xlabel("Janvier")
plt.ylabel("Février")
plt.savefig("janv vs fev")
```

### Histogrammes

```python
plt.hist(frame_salaries["TotalPay"],bins=50,color="red",label="Total",
        alpha=0.5)
plt.hist(frame_salaries["TotalPayBenefits"],bins=50,color="blue",
         label="TotalBenef",alpha=0.5)
plt.legend()
```

### Box plot

```python
plt.boxplot(frame_salaries["TotalPay"],notch=True)
```

Etude des données sur les logements en Californie

```python
frame_house=pd.read_csv("housing.csv")
```

```python
frame_house.info()
```

```python
#on représente les données
frame_house.plot("longitude","latitude",kind="scatter")
```

```python
#on affiche les données de position GPS avec la taille des communes et l'age médian des logements
plt.scatter("longitude","latitude",s=frame_house["population"]/100,
            c="housing_median_age",alpha=0.3,data=frame_house,
            label="population",cmap=plt.get_cmap("jet"))

plt.colorbar()
plt.legend()
```
